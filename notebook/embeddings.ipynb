```python
# Установка библиотек
!pip install -U sentence-transformers

# Импорты
import json
import numpy as np
from sentence_transformers import SentenceTransformer
from tqdm import tqdm

# Загрузка файла
from google.colab import files
uploaded = files.upload()  # загрузи файл cleaned_text_chunks.json

# Путь к файлу
INPUT_PATH = "cleaned_text_chunks.json"
EMBEDDINGS_PATH = "chunk_embeddings.npy"
IDS_PATH = "chunk_ids.json"

# Загружаем модель
print("Загружаем модель...")
model = SentenceTransformer("intfloat/multilingual-e5-base")

# Загружаем данные
with open(INPUT_PATH, encoding="utf-8") as f:
    chunks = json.load(f)

texts = [f"passage: {chunk['text']}" for chunk in chunks]
ids = [chunk['id'] for chunk in chunks]

# Генерация эмбеддингов
print("Генерация эмбеддингов...")
embeddings = model.encode(
    texts,
    show_progress_bar=True,
    batch_size=32,
    convert_to_numpy=True,
    normalize_embeddings=True
)

# Сохраняем
np.save(EMBEDDINGS_PATH, embeddings)
with open(IDS_PATH, "w", encoding="utf-8") as f:
    json.dump(ids, f, ensure_ascii=False, indent=2)

print("✅ Эмбеддинги успешно сгенерированы и сохранены.")

# Скачиваем результат
files.download(EMBEDDINGS_PATH)
files.download(IDS_PATH)
```
