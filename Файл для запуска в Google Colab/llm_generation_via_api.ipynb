```python
from flask import Flask, request, jsonify
from ctransformers import AutoModelForCausalLM

app = Flask(__name__)
print("üîÑ –ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª—å...")

model = AutoModelForCausalLM.from_pretrained(
    "/content/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    model_type="mistral",
    context_length=2048
)

@app.route('/generate', methods=['POST'])
def generate():
    data = request.get_json()
    prompt = data.get("prompt", "")
    print(f"üß† –ó–∞–ø—Ä–æ—Å: {prompt[:100]}...")
    output = model(prompt, max_new_tokens=100)
    return jsonify({"answer": output})
```

```python
from pyngrok import conf, ngrok


conf.get_default().auth_token = "–í–∞—à —Ç–æ–∫–µ–Ω"

public_url = ngrok.connect(7860)
print("üöÄ –ü—É–±–ª–∏—á–Ω—ã–π URL:", public_url)

with open("/content/llm_env.env", "w") as f:
    f.write(f"COLAB_LLM_URL={public_url}/generate\n")

app.run(host="0.0.0.0", port=7860)
```
